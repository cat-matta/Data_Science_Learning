<h1><center><span style="color:purple">A Background on Matplotlib</center></h1>
    
**Matplotlib is the most popular plotting module within Python. The module was originally developed by John Hunter, a PhD student studying neurobiology at the time. Originally Hunter was working across three different language, until he found Python. He found that Python was able to simplify much of his programming needs, but it still lagged behind Matlab in graphing and plotting. Hunter started developing the Matplotlib module to analyze human electrocorticography (ECoG) signals registered with 3D medical image data. From there it quickly grew to be the most popular plotting module in Python.**

<h1><center><span style="color:purple">A Background on seaborn</center></h1>

**seaborn is a plotting module for informative statistical graphs. It builds on top of Matplotlib and integrates closely with pandas data structures. Its plotting functions operate on dataframes and arrays containing whole datasets and internally perform the necessary semantic mapping and statistical aggregation to produce informative plots.**

<h1><center><span style="color:purple">A Background on Plotly</center></h1>

**Plotly is a module that was created by a group of friends who originally worked in Silicon Valley. The group came up with a program, Plot.ly, that allowed multiple users to collaborate on graphs and charts that required large amounts of data. One of the unique features of the Plotly module is that the graphs and visualizations are interactive. Once the graphs are generated, a user can hover a bar chart or scatter plot to view the information on the graph.**
